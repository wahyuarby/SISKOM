# Mengimpor library yang dibutuhkan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
```

# Memuat dataset dari file CSV

```python
data = pd.read_csv('/content/drive/MyDrive/SISKOM/complete laptop data0.csv', encoding='ISO-8859-1')

# Memilih kolom yang relevan
data = data[['name', 'Price', 'Processor Name', 'RAM', 'SSD Capacity']]
# Menampilkan data untuk memastikan formatnya sesuai
data.head()
```

# Preprocessing

```python
# Langkah 1: Menangani data kosong
# Cek apakah ada data yang kosong
print("Jumlah data kosong:\n", data.isnull().sum())

# Menghapus baris yang memiliki data kosong
# Karena kolom 'name' penting, kita pastikan kolom ini tidak kosong
data.dropna(subset=["name", "Price", "Processor Name", "RAM", "SSD Capacity"], inplace=True)

# Langkah 2: Membersihkan dan standarisasi data
# Menghapus simbol-simbol yang tidak diperlukan dari kolom 'Price'
data['Price'] = data['Price'].str.replace(r'[^\d]', '', regex=True).astype(int)  # Mengubah harga ke format numerik

# Kombinasi kolom menjadi satu fitur teks untuk TF-IDF
# Kita akan menggabungkan nama laptop, prosesor, RAM, dan SSD
# Hal ini digunakan untuk mencari kesamaan antar laptop
data['combined'] = data['name'] + ' ' + data['Processor Name'] + ' ' + data['RAM'] + ' ' + data['SSD Capacity']

# Langkah 3: Vektorisasi menggunakan TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(data['combined'])

# Menghitung kesamaan kosinus
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

# Fungsi untuk merekomendasikan laptop berdasarkan nama

```python
# Fungsi ini akan mengembalikan laptop yang mirip berdasarkan kesamaan konten
def recommend_laptops(title, data, cosine_sim, top_n=5):
    # Cari indeks laptop berdasarkan nama
    try:
        idx = data[data['name'] == title].index[0]
    except IndexError:
        return f"Laptop dengan nama '{title}' tidak ditemukan."

    # Ambil skor kesamaan pasangan
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Urutkan berdasarkan skor kesamaan secara menurun
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Ambil indeks laptop yang paling mirip
    sim_indices = [i[0] for i in sim_scores[1:top_n+1]]

    # Kembalikan nama laptop yang paling mirip
    return data.iloc[sim_indices][['name', 'Price', 'Processor Name', 'RAM', 'SSD Capacity']]

    # Contoh penggunaan fungsi rekomendasi
input_laptop = "Lenovo Lenovo V15 Celeron Dual Core - (4 GB/256 GB SSD/Windows 10) 82C30053IH Thin and Light Laptop  (15.6 inch, Iron Grey)"
recommendations = recommend_laptops(input_laptop, data, cosine_sim)

print(f"Rekomendasi untuk '{input_laptop}':")
print(recommendations)

```

# Menguji Rekomendasi

# Visualisasi

```python
# Visualisasi hasil rekomendasi
if isinstance(recommendations, pd.DataFrame):
    # Tabel hasil rekomendasi
    print("\nTabel Hasil Rekomendasi:")
    print(recommendations)

    # Grafik harga laptop yang direkomendasikan
    plt.figure(figsize=(10, 6))
    sns.barplot(x=recommendations['Price'], y=recommendations['name'], palette='viridis')
    plt.title(f"Harga Laptop yang Direkomendasikan untuk '{input_laptop}'")
    plt.xlabel('Harga (dalam satuan)')
    plt.ylabel('Nama Laptop')
    plt.tight_layout()
    plt.show()
```

# Menyimpan sistem rekomendasi ke file pickle

```python
class LaptopRecommender:
    def __init__(self, dataframe, content_col):
        self.df = dataframe
        self.content_col = content_col
        self.encoder = None
        self.bank = None

    def fit(self):
        self.encoder = TfidfVectorizer(stop_words='english')
        self.bank = self.encoder.fit_transform(self.df[self.content_col])

    def recommend(self, title, topk=5):
        try:
            idx = self.df[self.df['name'] == title].index[0]
            content = self.df.loc[idx, self.content_col]
            code = self.encoder.transform([content])
            dist = cosine_distances(code, self.bank)
            rec_idx = dist.argsort()[0, 1:(topk + 1)]
            return self.df.loc[rec_idx][['name', 'Price', 'Processor Name', 'RAM', 'SSD Capacity']]
        except IndexError:
            return f"Laptop '{title}' tidak ditemukan dalam dataset."

    def save_to_pickle(self, filename):
        with open(filename, 'wb') as f:
            pickle.dump(self, f)
        print(f"Sistem rekomendasi disimpan ke {filename}.")
```

# Membuat dan melatih sistem rekomendasi

```python
# Membuat dan melatih sistem rekomendasi
recsys = LaptopRecommender(data, content_col='combined')
recsys.fit()
```

```python
# Menyimpan model
recsys.save_to_pickle("laptop_recommender.pkl")
```

```python
# Memuat model dan menguji rekomendasi
with open("laptop_recommender.pkl", "rb") as f:
    loaded_recommender = pickle.load(f)
```

```python
loaded_recommendations = loaded_recommender.recommend(input_laptop)
print("Rekomendasi (dari model yang dimuat):")
print(loaded_recommendations)
```
